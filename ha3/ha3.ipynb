Here is $L_2(x)$:

$$
L_2(x) = \frac{(x-x_i)(x-x_{i+1})}{(x_{i-1}-x_i)(x_{i-1}-x_{i+1})}f_{i-1}
+ \frac{(x-x_{i-1})(x-x_{i+1})}{(x_i-x_{i-1})(x_i-x_{i+1})}f_i
+ \frac{(x-x_{i-1})(x-x_i)}{(x_{i+1}-x_{i-1})(x_{i+1}-x_i)}f_{i+1}
$$

Considering $h = const$ and $h = x_{i+1} - x_i = x_i - x_{i-1}$:

$$
L_2(x) = \frac{(x-x_i)}{-h}\frac{(x-x_{i+1})}{-2h}f_{i-1}
+ \frac{(x-x_{i-1})}{h}\frac{(x-x_{i+1})}{-h}f_i
+ \frac{(x-x_{i-1})}{2h}\frac{(x-x_i)}{h}f_{i+1}
$$

Now if we set $u = \frac{x - x_i}{h}$, observer that

$$
\frac{x - x_{i+1}}{h} = u - 1,\ \frac{x - x_{i-1}}{h} = u + 1
$$

$$
L_2(x) = \frac{u(u - 1)}{2}f_{i-1} - (u^2 - 1)f_i + \frac{(u + 1)u}{2}f_{i+1}
$$

Note that

$$
\frac{du}{dx} = \frac{1}{h} \implies dx = h \cdot du
$$

$$
\hat I_i^{i+2} = \int_i^{i+2} \bigg[\frac{u(u-1)}{2}f_i - (u^2-1)f_{i+1} + \frac{(u+1)u}{2}f_{i+2}\bigg]h \cdot du
$$

There is a subtle difference in this case as $u$ must be based upon middle point:

$$
u = \frac{x - x_{i+1}}{h}
$$

$$
u = 1,\ x = x_{i+2};\ u = -1,\ x = x_i
$$

$$
\hat I_i^{i+2} = h \bigg(\big(\frac{u^3}{6} - \frac{u^2}{4}\big)f_i
- \big(\frac{u^3}{3} - u\big)f_{i+1}
+ \big(\frac{u^3}{6} + \frac{u^2}{4}\big)f_{i+2}\bigg) \bigg|_i^{i+2}
$$

$$
\hat I_i^{i+2} = \frac{h}{3} (f_i + 4f_{i+1} + f_{i+2})
$$

It is handful to take $n = 2k$ during integration because then

$$
\hat I_b^a = \sum_{i=1}^k I_{2(i-1)}^{2i}
= \frac{h}{3}(f_0 + 4(f_1 + f_3 + \cdots + f_{n-1}) + 2(f_2 + f_4 + \cdots + f_{n-2}) + f_n)
$$

And we can cover the whole domain of integration $[a,b]$ using a single formula and we get a convenient formula
for $\hat I_b^a$ as a bonus. This happens because $L_2(x)$ uses three-point template and thus demands two sections per iteration. In case of $n = 2k + 1$ we will need to leave a single unpaired section, say at the end of integration domain and calculate its area separately. $I_{n-1}^n$ can be calculated with error of order $O(h^5)$,
using, for example, Euler-Maclaurin quadrature formula:

$$
\hat I_{n-1}^n = \frac{h}{24}(f_{n-3} - 5f_{n-2} + 19f_{n-1} + 9f_n)
$$

So in case of odd number of points $n = 2k + 1$ the formula is

$$
\hat I_b^a = \sum_{i=1}^k I_{2(i-1)}^{2i} + I_{n-1}^n
= \frac{h}{3}(f_0 + 4(f_1 + f_3 + \cdots + f_{n-2}) + 2(f_2 + f_4 + \cdots + f_{n-1}) + f_n)
+ \frac{h}{24}(f_{n-3} - 5f_{n-2} + 19f_{n-1} + 9f_n)
$$

Let us now estimate the error of integrating the function $f(x)$ using trapezoid method. Let us estimate the value
of the difference $I_i^{i+1} - \hat I_i^{i+1}$. According to Newton-Leibniz formula $I_i^{i+1} = F_{i+1} - F_i$ where $F_{i+1},\ F_i$ are antiderivatives of $f(x)$. Using the trapezoid method of integral approximation
$\hat I_i^{i+1} = \frac{h}{2}(f_i + f_{i+1})$ given $h = const$. So finally

$$
I_i^{i+1} - \hat I_i^{i+1} = F_{i+1} - F_i - \frac{h}{2}(f_i + f_{i+1})
$$

Now we can use Teylor series with the residual member in Lagrange form to approximate $F_{i+1} = F(x_{i+1})$. As
$x_{i+1} = x_i + h$,

$$
F_{i+1} = F(x_i) + f(x_i) h + f'(x_i) \frac{h^2}{2} + f''(x_i + \theta h) \frac{h^3}{6},
$$

where $\theta \in (0, 1)$. Let us say for simplicity that $\xi_i = x_i + \theta h$. Making analogous operation
with $f_{i+1} = f(x_{i+1})$ we get

$$
f_{i+1} = f(x_i) + f'(x_i) h + f''(\xi_i) \frac{h}{2}
$$

Substituting these into our difference formula yields

$$
I_i^{i+1} - \hat I_i^{i+1} = F_i + f_i h + f'(x_i) \frac{h^2}{2} + f''(\xi_i) \frac{h^3}{6} - F_i
- \frac{h}{2}\bigg(f_i + f_i + f'(x_i) h + f''(\xi_i) \frac{h^2}{2}\bigg)
$$

$$
I_i^{i+1} - \hat I_i^{i+1} = f_i h + f'(x_i) \frac{h^2}{2} + f''(\xi_i) \frac{h^3}{6}
- f_i h - \frac{h^2}{2} f'(x_i) - f''(\xi_i) \frac{h^3}{4}
$$

$$
I_i^{i+1} - \hat I_i^{i+1} = f''(\xi_i) \frac{h^3}{6} - f''(\xi_i) \frac{h^3}{4}
$$

$$
I_i^{i+1} - \hat I_i^{i+1} = - f''(\xi_i) \frac{h^3}{12}
$$

Then for $i \in [0,n]$

$$
\Big| I_i^{i+1} - \hat I_i^{i+1} \Big| \leq \max_{[x_i,x_{i+1}]} \big|f''(x)\big| \frac{h^3}{12}
$$

To estimate the error for $I_a^b - \hat I_a^b$ we need to accumulate the errors of all sections:

$$
I_a^b - \hat I_a^b = -\frac{h^3}{12} \sum_{i=0}^n f''(\xi_i) = -\frac{h^3}{12} \frac{b-a}{h} f''(\xi)
,\ \xi \in [a,b]
$$

$$
\Big| I_a^b - \hat I_a^b \Big| \leq \max_{[a,b]} \big|f''(x)\big| (b-a) \frac{h^2}{12}
$$
